# Import the libraries



```python
# Import the libraries
import numpy as np # To work with arrays
import pandas as pd # To work with the data
# Importing Sk learn libraries/ Modules
from sklearn.model_selection import train_test_split # Module to split the data into training and testing
from sklearn.linear_model import LogisticRegression # To build the model
from sklearn.metrics import accuracy_score # Evaluation
```

#Uploading CSV Data

```python
df=pd.read_csv("/heart_disease_data.csv")
```

```python
X = df.drop(columns='target',axis=1)
Y = df['target']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

#LogisticRegression Model

#

```python
model = LogisticRegression()
```

#Fitting the Model

```python
model.fit(X_train,Y_train)
```

#Training Data Prediction

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(f'This is the accuracy score on training data {training_data_accuracy}')
```

#Test Data Prediction

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(f'This is the accuracy score on training data {testing_data_accuracy}')
```

```python
training_data_accuracy-testing_data_accuracy
```
